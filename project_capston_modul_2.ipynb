```python
# Library

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore", FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, message="tight_layout not applied")
warnings.filterwarnings("ignore", category=UserWarning, message=".*Glyph 65281.*")
```

```python
# Import Data

airbnb = pd.read_csv("Airbnb Listings Bangkok.csv")
airbnb
```

```python
# Cek Info

airbnb.info()
```

```python
# Cek missing Value

missing_value = (airbnb.isnull().sum()/len(airbnb))*100
missing_value
```

```python
# Menghapus Missing Value <10%

threshold = 0.1

airbnb2 = airbnb.dropna(axis=1, thresh=int((threshold)*len(airbnb)))
airbnb2.info()
```

```python
# Visualisasi Heatmap

category = ['name','host_name','room_type','neighbourhood']
airbnb_dummies = pd.get_dummies(airbnb2, columns=category, drop_first=True)

# Menghitung korelasi antar kolom
correlation_matrix = airbnb_dummies.select_dtypes(include=['number']).corr()

# Membuat heatmap
plt.figure(figsize=(8, 8))  # Ukuran gambar
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral', fmt=".1f", square=True)

# Menampilkan heatmap
plt.title("Korelasi Antar Kolom setelah One-Hot Encoding")
plt.show()
```

```python
# Menghitung rata-rata sewa room 

average = airbnb2['minimum_nights'].mean()
average
```

```python
# Melihat room type yang paling banyak diminati

room = airbnb2['room_type'].value_counts()
room.plot(kind='bar')
plt.title('Tipe Kamar yang Paling Diminati')
plt.xlabel('Room Type')
plt.ylabel('Total Sewa')
```

```python
# Perbandingan rata-rata sewa dan room type

sns.barplot(x='room_type',y='price',data=airbnb2)
plt.title('Perbandingan Rata-rata Harga Sewa Berdasarkan Room Type')
plt.xlabel('Room Type')
plt.ylabel('Rata-rata Harga Sewa')
```

```python
# Review Customer terhadap masing-masing tipe room

sns.barplot(x='room_type',y='reviews_per_month',data=airbnb2)
plt.title('Banyaknya Review Berdasarkan Room Type')
plt.xlabel('Room Type')
plt.ylabel('Review Customer')
```

```python
# Membuat Plot dan Sub Plot
plt.figure(figsize=(12, 6))

# Plot pertama: Jumlah properti berdasarkan room_type
plt.subplot(1, 2, 1)
sns.countplot(x='room_type', data=airbnb, palette='viridis')
plt.title('Room Type yang Paling Diminati')
plt.xlabel('Room Type')
plt.ylabel('Jumlah Properti')

# Plot Kedua: Nama Apartement yang paling sering dipesan
ordered_top_5_apartments = airbnb2['name'].value_counts().index[:5]
plt.subplot(1, 3, 3)
sns.countplot(x='name', data=airbnb2[airbnb2['name'].isin(ordered_top_5_apartments)], palette='viridis', order=ordered_top_5_apartments)
plt.title('Jumlah Properti Yang Sering Dipesan Berdasarkan Nama Kamar')
plt.xlabel('Name')
plt.ylabel('Jumlah Properti')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Top 5 Apartment Yang Diminati Berdasarkan Neighborhood

apartment_neighbourhood = airbnb.groupby(['neighbourhood', 'name']).size().reset_index(name='total_bookings')
top_apartments_by_neighbourhood = apartment_neighbourhood.sort_values(by='total_bookings', ascending=False)
plt.figure(figsize=(12, 6))
top_5_apartments2 = top_apartments_by_neighbourhood.head(5)
sns.barplot(x='total_bookings', y='name', data=top_5_apartments2, palette='viridis')
plt.title('Top 5 Apartments Ordered in Neighbourhoods')
plt.xlabel('Total Bookings')
plt.ylabel('Apartment Name')
plt.show()
```

```python
# cek Korelasi Jumlah Review dan Tingkat Pemesanan

sns.scatterplot(x='number_of_reviews', y='reviews_per_month', data=airbnb)
plt.title('Hubungan Antara Jumlah Review dan Pemesanan')
plt.xlabel('Jumlah Review')
plt.ylabel('Jumlah Pemesanan per Bulan')
plt.show()
```

```python
# Cek Korelasi Harga dan Jumlah Review

sns.scatterplot(x='price', y='reviews_per_month', data=airbnb)
plt.title('Hubungan Antara Harga dan Jumlah Review')
plt.xlabel('Harga')
plt.ylabel('Jumlah Review')
plt.show()
```

```python

#Nama Apartement yang paling sering mendapatkan review

reviews_by_apartment = airbnb2.groupby('name')['reviews_per_month'].sum().sort_values(ascending=False)
top_5_apartments3 = reviews_by_apartment.head()
plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_apartments3.index, y=top_5_apartments3.values, palette='viridis')
plt.title('Top 5 Apartemen Berdasarkan Jumlah Review di Airbnb Bangkok')
plt.xlabel('Nama Apartemen')
plt.ylabel('Jumlah Review')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
